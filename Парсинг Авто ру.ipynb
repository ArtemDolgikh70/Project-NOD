```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
```

```python
chrome_options = Options()
#chrome_options.add_argument("--headless=new") 
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")
```

```python
driver = webdriver.Chrome(options=chrome_options)
driver.get("https://auto.ru/moskva/cars/all/")
```

```python
links = []
for i in driver.find_element(By.CLASS_NAME, "PageListing").find_elements(By.TAG_NAME, "a"):
    link = i.get_attribute("href")
    if link is None:
        continue
    if "sale" in link:
        links.append(link)
links = list(set(links))
```

```python
driver.get("https://auto.ru/cars/used/sale/bmw/x4/1132552839-e0fb7f5e/")
driver.find_element(By.CLASS_NAME, "CardHead__title").text.split(", ")[0]
```

```python
cars = {"model": [], "price": [], "owner": []}
for i in links:
    driver.get(i)
    print(i)
    model = driver.find_element(By.CLASS_NAME, "CardHead__title").text.split(", ")[0]
    cars["model"].append(model)
    price = driver.find_element(By.CLASS_NAME, "OfferPriceCaption__price").text
    cars["price"].append(price)
    owner = driver.find_element(By.CLASS_NAME, "CardOwner__ownerInfo").find_element(By.TAG_NAME, "a").get_attribute("title")
    cars["owner"].append(owner)
    ownership = driver.find_element(By.CLASS_NAME, "CardInfoSummary__list-jpQIS")
    for j in ownership.find_elements(By.TAG_NAME, "li"):
        ownership_data = j.text.split("\n")
        if ownership_data[0] not in cars:
            cars[ownership_data[0]] = [None] * (len(cars["price"]) - 1) + [ownership_data[1]]
        else:
            cars[ownership_data[0]].append(ownership_data[1])
    driver.find_element(By.XPATH, "//button[contains(., 'Все')]").click()
    for j in driver.find_element(By.CLASS_NAME, "ModificationInfo-iyGE4").find_elements(By.TAG_NAME, "li"):
        car_characteristics = j.text.split("\n")
        if len(car_characteristics) == 1:
            continue
        if car_characteristics[0] not in cars:
            cars[car_characteristics[0]] = [None] * (len(cars["price"]) - 1) + [car_characteristics[1]]
        else:
            cars[car_characteristics[0]].append(car_characteristics[1])
    for i in cars:
        if len(cars[i]) < len(cars["price"]):
            cars[i].append(None)
```
